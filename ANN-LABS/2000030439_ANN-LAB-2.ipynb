```python
import numpy as np
import matplotlib.pyplot as plt
import numpy as np
from math import e
```

```python
def binaryStep(x):
    ''' It returns '0' is the input is less then zero otherwise it returns one '''
    return np.heaviside(x,1)
```

```python
x = np.linspace(-10, 10)
plt.plot(x, binaryStep(x))
plt.axis('tight')
plt.title('Activation Function :binaryStep')
plt.show()
```

```python
def linear(x):
    ''' y = f(x) It returns the input as it is'''
    return x
```

```python
x = np.linspace(-10, 10)
plt.plot(x, linear(x))
plt.axis('tight')
plt.title('Activation Function :Linear')
plt.show()
```

# RELU Activation Function


```python
# Leaky Rectified Linear Unit (leaky ReLU) Activation Function
def leaky_ReLU(x):
 data = [max(0.01*value,value) for value in x]
 return np.array(data, dtype=float)
# Derivative for leaky ReLU 
def der_leaky_ReLU(x):
 data = [1 if value>0 else 0.01 for value in x]
 return np.array(data, dtype=float)
# Generating data For Graph
x_data = np.linspace(-10,10,100)
y_data = leaky_ReLU(x_data)
dy_data = der_leaky_ReLU(x_data)
# Graph
plt.plot(x_data, y_data, x_data, dy_data)
plt.title('leaky ReLU Activation Function & Derivative')
plt.legend(['leaky_ReLU','der_leaky_ReLU'])
plt.grid()
plt.show()
```

# Sigmoid or logistic activation function 

```python
def sigmoid(x):
    ''' It returns 1/(1+exp(-x)). where the values lies between zero and one '''

    return 1/(1+np.exp(-x))
```

```python
x = np.linspace(-10, 10)
plt.plot(x, sigmoid(x))
plt.axis('tight')
plt.title('Activation Function :Sigmoid')
plt.show()
```

# Hyperbolic tan activation function

```python
# Hyperbolic Tangent (htan) Activation Function
def htan(x):
 return (np.exp(x) - np.exp(-x))/(np.exp(x) + np.exp(-x))
# htan derivative
def der_htan(x):
 return 1 - htan(x) * htan(x)
# Generating data for Graph
x_data = np.linspace(-6,6,100)
y_data = htan(x_data)
dy_data = der_htan(x_data)
# Graph
plt.plot(x_data, y_data, x_data, dy_data)
plt.title('htan Activation Function & Derivative')
plt.legend(['htan','der_htan'])
plt.grid()
plt.show()
```

#  Bipolar sigmoid activation function

```python
def bipolarsigmoid(x):
 return (1-np.exp(-x))/(1+np.exp(-x))
x = np.linspace(-10, 10)
plt.plot(x, sigmoid(x))
plt.axis('tight')
plt.title('Activation Function :Sigmoid')
plt.show()
```

# Softplus activation function

```python
def softplus(x):
 return np.log(1+((e)**x))
# SoftPlus derivative: 
def softplus_deriv(x):
 return 1/(1+((e)**-x))
#generating the graph data
x_data = np.linspace(-10,10,100)
y_data = softplus(x_data)
dy_data = softplus_deriv(x_data)
#graph
plt.plot(x_data, y_data, x_data, dy_data)
plt.title('softplus Activation Function & Derivative')
plt.legend(['softplus','softplus_deriv'])
plt.grid()
plt.show()
```

# leaky RELU


```python
import numpy as np
import matplotlib.pyplot as plt

# Leaky Rectified Linear Unit (leaky ReLU) Activation Function
def leaky_ReLU(x):
  data = [max(0.05*value,value) for value in x]
  return np.array(data, dtype=float)

# Derivative for leaky ReLU 
def der_leaky_ReLU(x):
  data = [1 if value>0 else 0.05 for value in x]
  return np.array(data, dtype=float)

# Generating data For Graph
x_data = np.linspace(-10,10,100)
y_data = leaky_ReLU(x_data)
dy_data = der_leaky_ReLU(x_data)

# Graph
plt.plot(x_data, y_data, x_data, dy_data)
plt.title('leaky ReLU Activation Function & Derivative')
plt.legend(['leaky_ReLU','der_leaky_ReLU'])
plt.grid()
plt.show()
```
